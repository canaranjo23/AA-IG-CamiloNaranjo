```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Creamos los vector y las matrices de los parametros a y b 
ai = np.arange(0.1, 6)     # Vector de medias 
bi = np.arange(0.1, 6)     # Vector de varianzas

ai[1:] = np.round(ai[1:])
bi[1:] = np.round(bi[1:])
am, bm = np.meshgrid(ai,bi) 
print('a: ',am)
print('b: ',bm)

x = np.linspace(-5, 6,200)
print(len(x))
y = gamma(x)

# Plot 
fig = plt.figure(dpi=100)
plt.plot(x, y, 'b', alpha=0.6, label='gamma(x)')
plt.xlabel(r'$\lambda$')
plt.ylabel('Gamma function values')
plt.grid(True, linestyle='--')
plt.legend()
```

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Parametros
bi = [0.5, 1.0, 1.5,2.0]
loc = 0              # Determina donde comienza la grafica

# Plot 
fig, (ax1,ax2,ax3) = plt.subplots(3, 1, figsize=(5.0,9.0), 
                                  dpi=100, constrained_layout=True)

ai = 0.5
x = np.linspace(loc, 3, 200)
for j in range(0,len(bi)):
  lbl = 'a='+str(ai)+r', $\theta$='+str(bi[j])
  scale = bi[j]
  ax1.plot(x, gamma.pdf(x, ai, loc, scale), label=lbl)

ax1.set_xlabel(r'$\lambda$')
ax1.set_ylabel('Probability Density')
ax1.grid(True, linestyle='--')
ax1.legend()
ax1.text(-0.07, 6.2, 'a', fontsize=12, fontweight='bold', ha='center')

ai = 1.0
x = np.linspace(loc, 10, 200)
for j in range(0,len(bi)):
  lbl = 'a='+str(ai)+r', $\theta$='+str(bi[j])
  scale = bi[j]
  ax2.plot(x, gamma.pdf(x, ai, loc, scale), label=lbl)

ax2.set_xlabel(r'$\lambda$')
ax2.set_ylabel('Probability Density')
ax2.grid(True, linestyle='--')
ax2.legend()
ax2.text(-0.25, 1.9, 'b', fontsize=12, fontweight='bold', ha='center')

ai = 2.0
x = np.linspace(loc, 10, 200)
for j in range(0,len(bi)):
  lbl = 'a='+str(ai)+r', $\theta$='+str(bi[j])
  scale = bi[j]
  ax3.plot(x, gamma.pdf(x, ai, loc, scale), label=lbl)

ax3.set_xlabel(r'$\lambda$')
ax3.set_ylabel('Probability Density')
ax3.grid(True, linestyle='--')
ax3.legend()
ax3.text(-0.25, 0.72, 'c', fontsize=12, fontweight='bold', ha='center')
```

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
from scipy.stats import poisson

# Plot 
fig = plt.figure(dpi=100)

x = np.linspace(0, 10, 11)
a = 2.0
theta = 2.0

  #Poisson
plt.plot(x, poisson.pmf(x, 2.0), 'go', label='Poisson', zorder=4)
plt.vlines(x, 0, poisson.pmf(x, 2.0), colors='g', lw=5, alpha=0.5, zorder=4)

x = np.linspace(0, 10, 200)
  #Gamma
plt.plot(x, gamma.pdf(x, a, 0, theta), 'k', linewidth=8, label='Gamma')

  #Erlang
plt.plot(x, gamma.pdf(x, a, 0, theta), 'r', label='Erlang')

  #Chi-squared
plt.plot(x, gamma.pdf(x, a/2, 0, 2), 'b--', label='Chi-Squared')

plt.xlabel(r'$\lambda$')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--')
plt.legend()
```
